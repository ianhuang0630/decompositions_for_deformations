```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
surf_samples = np.load("debug/test_surf.npy")
bbox_samples = np.load("debug/test_bbox.npy")
surf_coord = surf_samples[:, :3]
surf_distance = surf_samples[:, 3]
bbox_coord = bbox_samples[:, :3]
bbox_distance = bbox_samples[:, 3]
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(bbox_coord[:, 0], bbox_coord[:, 1], bbox_coord[:, 2], )
ax.scatter3D(surf_coord[:, 0], surf_coord[:, 1], surf_coord[:, 2], c="red")


all_samples = np.vstack((bbox_samples, surf_samples))
all_coords = all_samples[:, :3]
all_dist = all_samples[:, 3]

inside_coord = all_coords[np.where(all_dist > 0)[0], :]
outside_coord = all_coords[np.where(all_dist <= 0)[0], :]
print("inside: {}".format(len(inside_coord)))
print("outside: {}".format(len(outside_coord)))

for i in range(4):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.view_init(elev=32., azim=(i+1)*45)
    # ax.scatter3D(outside_coord[:, 0], outside_coord[:, 1], outside_coord[:, 2])

    ax.scatter3D(inside_coord[:, 0], inside_coord[:, 1], inside_coord[:, 2], c="red")
```
